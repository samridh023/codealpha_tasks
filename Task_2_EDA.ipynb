# Task 2: Exploratory Data Analysis (EDA)

**Dataset:** Students Performance Dataset

**Objective:** Analyze student performance and identify patterns using EDA.

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset (sample structure)
data = {
    'math score': [72, 69, 90, 47, 76],
    'reading score': [72, 90, 95, 57, 78],
    'writing score': [74, 88, 93, 44, 75],
    'test preparation course': ['none','completed','none','none','completed']
}
df = pd.DataFrame(data)
df.head()
```

## Dataset Information

```
df.info()
```

## Statistical Summary

```
df.describe()
```

## Missing Values

```
df.isnull().sum()
```

## Math Score Distribution

```
plt.figure()
sns.histplot(df['math score'], kde=True)
plt.show()
```

## Test Preparation vs Math Score

```
plt.figure()
sns.boxplot(x='test preparation course', y='math score', data=df)
plt.show()
```

## Correlation Analysis

```
plt.figure()
sns.heatmap(df[['math score','reading score','writing score']].corr(), annot=True)
plt.show()
```

## Conclusion
Test preparation improves performance and strong correlation exists between reading and writing scores.
