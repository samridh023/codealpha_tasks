# Task 3: Data Visualization

Objective: Visualize data using different plotting techniques and extract insights.


```
import pandas as pd
import matplotlib.pyplot as plt

# Load sample dataset (Student Performance)
data = {
    'gender': ['male','female','female','male','female','male','female','male'],
    'math_score': [78, 88, 90, 65, 72, 85, 91, 69],
    'reading_score': [75, 92, 95, 60, 70, 88, 93, 68],
    'writing_score': [70, 94, 96, 58, 68, 90, 95, 66]
}

df = pd.DataFrame(data)
df
```

## Bar Chart
Gender distribution of students.

```
plt.figure()
df['gender'].value_counts().plot(kind='bar')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

**Insight:** Female students are slightly more represented in the dataset.

## Histogram
Distribution of Math Scores.

```
plt.figure()
plt.hist(df['math_score'], bins=5)
plt.title('Math Score Distribution')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.show()
```

**Insight:** Most students scored between 70 and 90 in math.

## Box Plot
Comparison of Math Scores by Gender.

```
plt.figure()
df.boxplot(column='math_score', by='gender')
plt.title('Math Scores by Gender')
plt.suptitle('')
plt.xlabel('Gender')
plt.ylabel('Math Score')
plt.show()
```

**Insight:** Female students show slightly higher median math scores.

## Line Chart
Average scores by subject.

```
plt.figure()
avg_scores = df[['math_score','reading_score','writing_score']].mean()
avg_scores.plot(kind='line', marker='o')
plt.title('Average Scores by Subject')
plt.ylabel('Average Score')
plt.show()
```

**Insight:** Writing scores are highest on average.

## Heatmap
Correlation between scores.

```
plt.figure()
corr = df[['math_score','reading_score','writing_score']].corr()
plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title('Score Correlation Heatmap')
plt.show()
```

**Insight:** Strong positive correlation exists among all subject scores.
